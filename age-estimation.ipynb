```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.utils.np_utils import to_categorical
from keras.layers import Dense, Dropout, Flatten, Input
from keras.applications.vgg16 import VGG16
from keras.models import Model, Sequential
from keras import optimizers
from os import listdir


# データリストの取得

path_age0 = [filename for filename in listdir('./age0-9') if not filename.startswith('.')]
path_age1 = [filename for filename in listdir('./age10-19') if not filename.startswith('.')]
path_age2 = [filename for filename in listdir('./age20-29') if not filename.startswith('.')]
path_age3 = [filename for filename in listdir('./age30-39') if not filename.startswith('.')]
path_age4 = [filename for filename in listdir('./age40-49') if not filename.startswith('.')]
path_age5 = [filename for filename in listdir('./age50-59') if not filename.startswith('.')]
path_age6 = [filename for filename in listdir('./age60-') if not filename.startswith('.')]


#画像データの読み込み
img_age0 = []
img_age2 = []


for i in range(len(path_age0)):
    img = cv2.imread('./age0-9/' + path_age0[i])
    img = cv2.resize(img, (150,150))
    img_age0.append(img)

for i in range(len(path_age1)):
    img = cv2.imread('./age10-19/' + path_age1[i])
    img = cv2.resize(img, (150,150))
    img_age0.append(img)

for i in range(len(path_age2)):
    img = cv2.imread('./age20-29/' + path_age2[i])
    img = cv2.resize(img, (150,150))
    img_age2.append(img)

for i in range(len(path_age3)):
    img = cv2.imread('./age30-39/' + path_age3[i])
    img = cv2.resize(img, (150,150))
    img_age2.append(img)

for i in range(len(path_age4)):
    img = cv2.imread('./age40-49/' + path_age4[i])
    img = cv2.resize(img, (150,150))
    img_age2.append(img)

for i in range(len(path_age5)):
    img = cv2.imread('./age50-59/' + path_age5[i])
    img = cv2.resize(img, (150,150))
    img_age2.append(img)

for i in range(len(path_age6)):
    img = cv2.imread('./age60-/' + path_age6[i])
    img = cv2.resize(img, (150,150))
    img_age2.append(img)

    
X = np.array(img_age0 + img_age2)
y =  np.array([0]*len(img_age0) + [1]*len(img_age2))

rand_index = np.random.permutation(np.arange(len(X)))
X = X[rand_index]
y = y[rand_index]

# データの分割
X_train = X[:int(len(X)*0.8)]
y_train = y[:int(len(y)*0.8)]
X_test = X[int(len(X)*0.8):]
y_test = y[int(len(y)*0.8):]


y_train = to_categorical(y_train)
y_test = to_categorical(y_test)


input_tensor = Input(shape = (150, 150, 3))
vgg16 = VGG16(include_top = False, weights = "imagenet", input_tensor = input_tensor)

top_model = vgg16.output
top_model = Flatten(input_shape=vgg16.output_shape[1:])(top_model)
top_model = Dense(128, activation='relu')(top_model)
top_model = Dense(128, activation="relu")(top_model)
top_model = Dropout(0.5)(top_model)
top_model = Dense(2, activation='softmax')(top_model)

model = Model(inputs=vgg16.input, outputs=top_model)

for layer in model.layers[:15]:
    layer.trainable = False

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.SGD(lr=1e-5, momentum=0.9),
              metrics=['accuracy'])

history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=64, epochs=20)

model.save('age_model0709.h5')#重みを保存する実装が必要

# 精度の評価
scores = model.evaluate(X_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```

```python
#acc, val_accのプロット
plt.plot(history.history["acc"], label="acc", ls="-", marker="o")
plt.plot(history.history["val_acc"], label="val_acc", ls="-", marker="x")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(loc="best")
plt.show()
#loss, val_lossのプロット
plt.plot(history.history["loss"], label="loss", ls="-", marker="o")
plt.plot(history.history["val_loss"], label="val_loss", ls="-", marker="x")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(loc="best")
plt.show()
```
