```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.utils.np_utils import to_categorical
from keras.layers import Dense, Dropout, Flatten, Input
from keras.applications.vgg16 import VGG16
from keras.models import Model, Sequential
from keras import optimizers
from os import listdir


# データリストの取得
path_male0 = [filename for filename in listdir('./male/male0-9') if not filename.startswith('.')]
path_male1 = [filename for filename in listdir('./male/male10-19') if not filename.startswith('.')]
path_male2 = [filename for filename in listdir('./male/male20-29') if not filename.startswith('.')]
path_male3 = [filename for filename in listdir('./male/male30-39') if not filename.startswith('.')]
path_male4 = [filename for filename in listdir('./male/male40-49') if not filename.startswith('.')]
path_male5 = [filename for filename in listdir('./male/male50-59') if not filename.startswith('.')]
path_male6 = [filename for filename in listdir('./male/male60-') if not filename.startswith('.')]

path_female0 = [filename for filename in listdir('./female/female0-9') if not filename.startswith('.')]
path_female1 = [filename for filename in listdir('./female/female10-19') if not filename.startswith('.')]
path_female2 = [filename for filename in listdir('./female/female20-29') if not filename.startswith('.')]
path_female3 = [filename for filename in listdir('./female/female30-39') if not filename.startswith('.')]
path_female4 = [filename for filename in listdir('./female/female40-49') if not filename.startswith('.')]
path_female5 = [filename for filename in listdir('./female/female50-59') if not filename.startswith('.')]
path_female6 = [filename for filename in listdir('./female/female60-') if not filename.startswith('.')]

#画像データの読み込み
img_male = []
img_female = []


for i in range(len(path_male0)):
    img = cv2.imread('./male/male0-9/' + path_male0[i])
    img = cv2.resize(img, (150,150))
    img_male.append(img)

for i in range(len(path_male1)):
    img = cv2.imread('./male/male10-19/' + path_male1[i])
    img = cv2.resize(img, (150,150))
    img_male.append(img)

for i in range(len(path_male2)):
    img = cv2.imread('./male/male20-29/' + path_male2[i])
    img = cv2.resize(img, (150,150))
    img_male.append(img)

for i in range(len(path_male3)):
    img = cv2.imread('./male/male30-39/' + path_male3[i])
    img = cv2.resize(img, (150,150))
    img_male.append(img)

for i in range(len(path_male4)):
    img = cv2.imread('./male/male40-49/' + path_male4[i])
    img = cv2.resize(img, (150,150))
    img_male.append(img)
    
for i in range(len(path_male5)):
    img = cv2.imread('./male/male50-59/' + path_male5[i])
    img = cv2.resize(img, (150,150))
    img_male.append(img)
    
for i in range(len(path_male6)):
    img = cv2.imread('./male/male60-/' + path_male6[i])
    img = cv2.resize(img, (150,150))
    img_male.append(img)

for i in range(len(path_female0)):
    img = cv2.imread('./female/female0-9/' + path_female0[i])
    img = cv2.resize(img, (150,150))
    img_female.append(img)
    
for i in range(len(path_female1)):
    img = cv2.imread('./female/female10-19/' + path_female1[i])
    img = cv2.resize(img, (150,150))
    img_female.append(img)

for i in range(len(path_female2)):
    img = cv2.imread('./female/female20-29/' + path_female2[i])
    img = cv2.resize(img, (150,150))
    img_female.append(img)
    
for i in range(len(path_female3)):
    img = cv2.imread('./female/female30-39/' + path_female3[i])
    img = cv2.resize(img, (150,150))
    img_female.append(img)
    
for i in range(len(path_female4)):
    img = cv2.imread('./female/female40-49/' + path_female4[i])
    img = cv2.resize(img, (150,150))
    img_female.append(img)
    
for i in range(len(path_female5)):
    img = cv2.imread('./female/female50-59/' + path_female5[i])
    img = cv2.resize(img, (150,150))
    img_female.append(img)
    
for i in range(len(path_female6)):
    img = cv2.imread('./female/female60-/' + path_female6[i])
    img = cv2.resize(img, (150,150))
    img_female.append(img)


X = np.array(img_male + img_female)
y =  np.array([0]*len(img_male) + [1]*len(img_female))

rand_index = np.random.permutation(np.arange(len(X)))
X = X[rand_index]
y = y[rand_index]

# データの分割
X_train = X[:int(len(X)*0.8)]
y_train = y[:int(len(y)*0.8)]
X_test = X[int(len(X)*0.8):]
y_test = y[int(len(y)*0.8):]


y_train = to_categorical(y_train)
y_test = to_categorical(y_test)


input_tensor = Input(shape = (150, 150, 3))
vgg16 = VGG16(include_top = False, weights = "imagenet", input_tensor = input_tensor)

top_model = vgg16.output
top_model = Flatten(input_shape=vgg16.output_shape[1:])(top_model)
top_model = Dense(64, activation='relu')(top_model)
top_model = Dropout(0.5)(top_model)
top_model = Dense(128, activation="relu")(top_model)
top_model = Dense(2, activation='softmax')(top_model)

model = Model(inputs=vgg16.input, outputs=top_model)

for layer in model.layers[:16]:
    layer.trainable = False

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.SGD(lr=1e-5, momentum=0.9),
              metrics=['accuracy'])

history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=64, epochs=20)

#重みを保存する実装が必要
model.save('gender_model0709.h5')
        
# 精度の評価
scores = model.evaluate(X_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```

```python
#acc, val_accのプロット
plt.plot(history.history["accuracy"], label="acc", ls="-", marker="o")
plt.plot(history.history["val_accuracy"], label="val_acc", ls="-", marker="x")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(loc="best")
plt.show()
#loss, val_lossのプロット
plt.plot(history.history["loss"], label="loss", ls="-", marker="o")
plt.plot(history.history["val_loss"], label="val_loss", ls="-", marker="x")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(loc="best")
plt.show()
```
